```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import glob
import os
```

```python
matches_df = pd.read_csv(r'E:\EDA\ICC Cricket World Cup\matches.csv')
matches_df
```

There are some un named columns. First lets remove them.

```python
# Get a list of columns containing 'Unnamed'
unnamed_columns = [col for col in matches_df.columns if 'Unnamed' in col]
print(unnamed_columns)
matches_df.drop(columns=unnamed_columns, inplace=True)
matches_df
```

```python
# Check for missing values
missing_values = matches_df.isnull().sum()

# Calculate missing value percentages
missing_percentage = (missing_values / len(matches_df)) * 100

# Create a DataFrame to display missing values and percentages
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
print(missing_data)
```

```python
# Fill missing values in toss.elected and toss.winner columns with 'N/A'
matches_df['toss.elected'].fillna('N/A', inplace=True)
matches_df['toss.winner'].fillna('N/A', inplace=True)
```

```python
matches_df.info()
```

Chnage the data type of matchDate, matchEndDate to DateTime from Object

```python
matches_df['matchDate'] = pd.to_datetime(matches_df['matchDate'], format="%Y-%m-%dT%H:%M:%S%z", utc=True, errors='coerce')
matches_df['matchEndDate'] = pd.to_datetime(matches_df['matchEndDate'], format="%Y-%m-%dT%H:%M:%S%z", utc=True, errors='coerce')
```

Lets add another column 'Year' after matchEndDate column. It might be used in the future.

```python
# Extract year from 'matchEndDate' column
matches_df['Year'] = matches_df['matchEndDate'].dt.year

# Reorder columns to place 'Year' column after 'matchEndDate'
columns = list(matches_df.columns)
match_end_date_index = columns.index('matchEndDate')

# Check if 'Year' column already exists and remove if present elsewhere
if 'Year' in columns and columns[-1] != 'Year':
    columns.remove('Year')

# Remove 'Year' column from its previous position if found
if 'Year' in columns and match_end_date_index != len(columns) - 1:
    columns.remove('Year')

# Insert 'Year' column after 'matchEndDate'
columns.insert(match_end_date_index + 1, 'Year')
```

```python
#Lets fill the missing values in matchStatus_victoryMarginRuns and matchStatus_victoryMarginWickets based on matchStatus_text
def update_victory_margin(row):
    text = str(row['matchStatus_text']).lower()  # Convert text to lowercase for uniform comparison
    runs_match = re.search(r'(\d+) runs', text)
    wickets_match = re.search(r'(\d+) wicket', text)

    if runs_match:
        runs = int(runs_match.group(1))  # Extract runs
        return runs, np.nan  # Fill 'matchStatus_victoryMarginRuns' with runs and 'matchStatus_victoryMarginWickets' as NaN
    elif wickets_match:
        wickets = int(wickets_match.group(1))  # Extract wickets
        return np.nan, wickets  # Fill 'matchStatus_victoryMarginRuns' as NaN and 'matchStatus_victoryMarginWickets' with wickets
    else:
        return np.nan, np.nan  # Fill both columns as NaN if no specific condition is met

# Apply the function to update 'matchStatus_victoryMarginRuns' and 'matchStatus_victoryMarginWickets'
matches_df[['matchStatus_victoryMarginRuns', 'matchStatus_victoryMarginWickets']] = matches_df.apply(update_victory_margin, axis=1, result_type='expand')
```

Lets make another column named "Winner" which display the name of the winning team

```python
def extract_winner(text):
    words = text.split(' ')
    if words[0].isalpha():  
        return words[0]  
    else:
        return text  

# Create the 'Winner' column by extracting country names or using the entire text
matches_df['Winner'] = matches_df['matchStatus_text'].apply(extract_winner)
```

```python
# Removing duplicates based on all columns
matches_df.drop_duplicates(inplace=True)
```
