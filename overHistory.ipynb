```python
import pandas as pd
import numpy as np
import re
import os
import glob
import matplotlib.pyplot as plt
```

```python
overHistory_df = pd.read_csv('E:\EDA\ICC Cricket World Cup\overHistory.csv')
overHistory_df
```

```python
# List of columns to be deleted
columns_to_delete = ['ovBalls/7', 'ovBalls/8', 'ovBalls/9','ovBalls/10','ovBalls/11','ovBalls/12','ovBalls/13']

# Dropping specified columns
overHistory_df.drop(columns=columns_to_delete, inplace=True)
```

```python
overHistory_df.info()
```

There is '.' in ovBalls columns denoting the dot ball. we will replace this with 0.

```python
# Replace '.' with '0' in the specified columns
columns_to_clean = ['ovBalls/0', 'ovBalls/1', 'ovBalls/2', 'ovBalls/3', 'ovBalls/4', 'ovBalls/5', 'ovBalls/6']

for column in columns_to_clean:
    overHistory_df[column] = overHistory_df[column].replace('.', '0')

# Fill missing values with '0' in the specified columns
overHistory_df[columns_to_clean] = overHistory_df[columns_to_clean].fillna('0')
```

```python
overHistory_df.head(5)
```

```python
# List of columns to be deleted
columns_to_delete1 = ['bowlerId']

# Dropping specified columns
overHistory_df.drop(columns=columns_to_delete1, inplace=True)
```

Add another column totalRuns which calculate total runs in an over.

```python
# Function to calculate total runs
def calculate_total(row):
    total = 0
    for val in row.values:
        num = re.findall(r'\d+', str(val))  # Extract numerical values using regex
        if num:
            total += int(num[0])  # Add the numeric value to total
    return total

# Columns to consider for calculation
columns_to_consider = [col for col in overHistory_df.columns if col.startswith('ovBalls/')]

# Apply the function across specified columns to calculate 'totalRuns' column
overHistory_df['totalRuns'] = overHistory_df[columns_to_consider].apply(calculate_total, axis=1)
```
